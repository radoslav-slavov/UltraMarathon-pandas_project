```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
```

```python
df.dtypes
```

```python
#clean up data
```

```python
#only want USA races 50k or 50mi, 2020
```

```python
#step 1 show 50mi or 50km
#50km
#50mi
```

```python
df[df['Event distance/length'] == '50mi']
```

```python
#combine 50km and 50mi with isin
```

```python
df[df['Event distance/length'].isin(['50km','50mi']) & (df['Year of event'] == 2020)]
```

```python
df[df['Event name'] == 'Everglades 50 Mile Ultra Run (USA)']['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)
```

```python
#return all USA races
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
#combine all filters together
```

```python
df[df['Event distance/length'].isin(['50km','50mi']) &
(df['Year of event'] == 2020) & 
(df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')] 
```

```python
df2 = df[df['Event distance/length'].isin(['50km','50mi']) &
(df['Year of event'] == 2020) & 
(df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head()
```

```python
df2.shape
```

```python
#remove (USA) from event name
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
#clean up athlete birth year from float to int and replace nulls with 0
```

```python
df2['Athlete year of birth'] = df2['Athlete year of birth'].fillna(0)
```

```python
df2['Athlete year of birth'] = df2['Athlete year of birth'].astype(int)
```

```python
df2['Athlete age'] = 2020 - df2['Athlete year of birth']
```

```python
#remove h from athlete perfomance
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.replace(' h', '')
```

```python
df2.head(10)
```

```python
#drop columns: Athlete Club, Athlete Country, Athlete year of birth, Athlete age category
```

```python
df2.head(10)
```

```python
#check for duplicates
```

```python
df2[df2.duplicated() == True]
```

```python
#reset index
```

```python
df2.reset_index(drop=True)
```

```python
#fix types
```

```python
df2.dtypes
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
#rename columns
```

```python
#Year of event                  int64
#Event dates                   object
#Event name                    object
#Event distance/length         object
#Event number of finishers      int64
#Athlete performance           object
#Athlete gender                object
#Athlete average speed        float64
#Athlete ID                     int64
#Athlete age                    int32
```

```python
df2 = df2.rename(columns = {'Year of event': 'year',
                            'Event dates': 'race_day',
                            'Event name': 'race_name',
                            'Event distance/length': 'race_length',
                            'Event number of finishers': 'race_number_of_finishers',
                            'Athlete performance': 'athlete_performance',
                            'Athlete gender': 'athlete_gender',
                            'Athlete average speed': 'athlete_average_speed',
                            'Athlete ID': 'athlete_id',
                            'Athlete age': 'athlete_age'
                           })
```

```python
df2.head()
```

```python
#reorder columns
```

```python
df3 = df2[['race_day', 'race_name', 'race_length', 'race_number_of_finishers', 'athlete_id', 'athlete_gender', 'athlete_age', 'athlete_performance', 'athlete_average_speed']]
```

```python
df3 = df3[df3['athlete_age'] < 100]
```

```python
df3.head()
```

```python
sns.histplot(df3['race_length'])
```

```python
sns.histplot(df3, x= 'race_length', hue = 'athlete_gender')
```

```python
sns.displot(df3[df3['race_length'] == '50mi']['athlete_average_speed'])
```

```python
sns.violinplot(data = df3, x='race_length', y='athlete_average_speed', hue = 'athlete_gender', split = True, inner = 'quart')
```

```python
sns.lmplot(data=df3, x='athlete_age', y='athlete_average_speed', hue = 'athlete_gender')
```

```python
#questions to answer based on the prepared data
```

```python
#Difference in speed for 50km,50mi male to female 
```

```python
df3.groupby(['race_length','athlete_gender'])['athlete_average_speed'].mean()
```

```python
#what age groups are the best in the 50mi Race (minimum participation of 20+ races)
```

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False).query('count>19')
```

```python
#what age groups are the worst in the 50mi Race (minimum participation of 10+ races)
```

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = True).query('count>9')
```
